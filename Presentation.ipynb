<h1> Some methods for Solar data analysis in high resolution: </h1>
</center>


- Pre-processing data:
  - Speckle (Statistical reconstruction for the most likely intensity map)
  - MOMBFD (Statistical recontruction from the Spatial frequencies)
  - Bigsonic (Low-pass band to remove p-modes in photospheric maps)
  - Coaligment (Maps alignment to correct solar differential rotation and proper displacements of features)
  - Wavelength calibrations and so  on!

### Processing data

- Proper motions dynamics:
  - Local Correlation Tracking (LCT, and its derivations)
  - Differential Affine Velocity Estimator for Vector Magnetograms (DAVE4VM)
  - Deep learning for the estimation of horizontal velocities at the solar surface(DeepVel)
  - and many more.

- Inversion of Stokes profiles
  - Stokes Inversion based on Response functions (SIR)
  - HAnle and ZEeman Light (HAZEL)
  - Non-LTE Inversion COde using Lorien Engine (NICOLE)
  - The Milne-Eddington gRid Linear Inversion Network (MERLIN) 
  - MILOS and many others.

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import hazel
import h5py
import scipy.io as io
print(hazel.__version__)
label = ['I', 'Q', 'U', 'V']
````

````python
tmp = h5py.File('../GREGOR/Test_Hazel/Full_data.h5py','r')

Stokes = tmp['Data']
wavelength = tmp['Wave Lenght']

stokes_t = np.zeros((4,355))
stokes_t[0,:] = Stokes['Stokes I'][216,:,16]
stokes_t[1,:] = Stokes['Stokes Q'][216,:,16]
stokes_t[2,:] = Stokes['Stokes U'][216,:,16]
stokes_t[3,:] = Stokes['Stokes V'][216,:,16]

dlambda = wavelength['dlambda']
lambdac = wavelength['lambdac']
````

````python
f, ax = plt.subplots(nrows=1, ncols=2, figsize=(14,9))
ax[0].imshow(Stokes['Stokes I'][:,0,:],origin='lower',cmap='gray')
ax[1].imshow(Stokes['Stokes I'][:,300,:],origin='lower',cmap='hot')
ax[0].plot(16,216,'r+')
ax[1].plot(16,216,'b+')
````

````python
f, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
ax[0,0].plot(lambdac, stokes_t[0,:])
ax[0,1].plot(lambdac, stokes_t[1,:])
ax[1,0].plot(lambdac, stokes_t[2,:])
ax[1,1].plot(lambdac, stokes_t[3,:])
````

Now we compute a very rough estimation of the noise standard deviation, the weights, and observation files

````python
noise = np.std(stokes_t[1,0:20])

np.savetxt('10830_pore.wavelength', lambdac, header='lambda')

f = open('10830_pore.weights', 'w')
f.write('# WeightI WeightQ WeightU WeightV\n')
for i in range(len(lambdac)):
    f.write('1.0    1.0   1.0   1.0\n')
f.close()
stokes_t.shape

f = open('10830_pore_stokes_pix_16_216.1d', 'wb')
f.write(b'# LOS theta_LOS, phi_LOS, gamma_LOS\n')
f.write(b'0.0 0.0 90.0\n')
f.write(b'\n')
f.write(b'# Boundary condition I/Ic(mu=1), Q/Ic(mu=1), U/Ic(mu=1), V/Ic(mu=1)\n')
f.write(b'1.0 0.0 0.0 0.0\n')
f.write(b'\n')
f.write(b'# SI SQ SU SV sigmaI sigmaQ sigmaU sigmaV\n')
tmp2 = np.vstack([stokes_t, noise*np.ones((4,len(lambdac)))])
np.savetxt(f, tmp2.T)
f.close()
````

Now we are ready!!!! We can check carefully the configuration file for a while.. 🕵️‍♀️

````python
%cat ../GREGOR/Test_Hazel/conf_spot.ini
````

Then, we are really ready to make the inversion!!

````python
mod = hazel.Model('../GREGOR/Test_Hazel/conf_spot.ini', working_mode='inversion', verbose=2)
mod.read_observation()
mod.open_output()
mod.invert()
mod.write_output()
mod.close_output()
````
